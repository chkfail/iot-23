```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

```python

models = ['DRF','GBM','XGBoost','ANN']
accuracy = [0.996990,0.999270,0.999726,0.996808]
weighted_f1_avg = [0.996991,0.999270,0.999726,0.996809]

plt.figure(figsize=(8,6))
plt.scatter(accuracy, weighted_f1_avg, s=200, color="red")
plt.xlabel("Accuracy",fontsize=14)
plt.ylabel("Weighted F1 avg",fontsize=14)
plt.title("Zeek Dataset",fontsize=16)

for i, label in enumerate(models):
    plt.annotate(label, (accuracy[i], weighted_f1_avg[i]),fontsize=14)
plt.show()

```
